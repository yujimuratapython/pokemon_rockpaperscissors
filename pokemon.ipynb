```python
import requests
url = "https://pokeapi.co/api/v2/pokemon/pikachu/"

r = requests.get(url, timeout=5)
r = r.json()

name = r['name']
id = r['id']
image =r['sprites']['front_default']
types = r['types'][0]['type']['name']

print(id)
print(name)
print(image)
print(types)
```

```python
from tkinter import *
from random import randint
from tkinter import ttk


root = Tk()
root.title('hitokage, zenigame, fushigidane')
root.geometry("500x600")
# Change bg color to white
root.config(bg="#fcfcfc")

# Define our images
hitokage = PhotoImage(file='images/hitokage.png')
zenigame = PhotoImage(file='images/zenigame.png')
fushigidane = PhotoImage(file='images/fushigidane.png')

# Add Images to a list
image_list = [hitokage, zenigame, fushigidane]

# Pick random number between 0 and 2
pick_number = randint(0,2)

# Throw up an image when the program starts
image_label = Label(root, image=image_list[pick_number], bd=0)
image_label.pack(pady=20)

# Create Spin Function
def fight():
	# Pick random numnber
	pick_number = randint(0,2)
	# Show image
	image_label.config(image=image_list[pick_number])

	# 0 = hitokage
	# 1 = zenigame
	# 2 = fushigidane

	# Convert Dropdown choice to a number
	if user_choice.get() == "hitokage":
		user_choice_value = 0
	elif user_choice.get() == "zenigame":
		user_choice_value = 1
	elif user_choice.get() == "fushigidane":
		user_choice_value = 2


	# Determine if we won or lost
	if user_choice_value == 0: # hitokage
		if pick_number == 0:
			win_lose_label.config(text="ひきわけ")
		elif pick_number == 1: # zenigame
			win_lose_label.config(text="まけ")
		elif pick_number == 2: # fushigidane
			win_lose_label.config(text="かち")


	# If User Picks zenigame
	if user_choice_value == 1: # zenigame
		if pick_number == 1:
			win_lose_label.config(text="ひきわけ")
		elif pick_number == 0: # hitokage
			win_lose_label.config(text="かち")
		elif pick_number == 2: # fushigidane
			win_lose_label.config(text="まけ")

	# If User Pics fushigidane
	if user_choice_value == 2: # fushigidane
		if pick_number == 2:
			win_lose_label.config(text="ひきわけ")
		elif pick_number == 0: # hitokage
			win_lose_label.config(text="まけ")
		elif pick_number == 1: # zenigame
			win_lose_label.config(text="かち")


# Make our choice
user_choice = ttk.Combobox(root, value=("hitokage", "zenigame", "fushigidane"))
user_choice.current(0)
user_choice.pack(pady=20)

# Create Fight Button
fight_button = Button(root, text="Fight!", command=fight)
fight_button.pack(pady=10)

# Label for showing if you won or not
win_lose_label = Label(root, text="ポケモンを選んでね", font=("Helvetica", 18), bg="#fcfcfc")
win_lose_label.pack(pady=50)


root.mainloop()
```
